```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('images/someshapes.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_gray)
plt.show()
```

```python
ret,image_thresh=cv2.threshold(image_gray,127,255,1)
plt.imshow(image_thresh)
plt.show()
```

```python
edged_image = cv2.Canny(image_gray,0,255)
plt.imshow(edged_image)
plt.show()
```

```python
copy_image = edged_image.copy()
_, contours, hierarchy = cv2.findContours(copy_image, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
plt.imshow(copy_image)
plt.show()
```
