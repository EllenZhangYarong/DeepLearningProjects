```python
image_RGB = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_RGB)
plt.show()
```

```python
input = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(input, cmap='gray')
plt.show()
```

```python
# Sharpen
kernel_sharpening = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
sharpened_image = cv2.filter2D(input,-1,kernel_sharpening)
plt.imshow(sharpened_image, cmap='gray')
plt.show()
```

```python
# Blur
bilateral_image = cv2.bilateralFilter(input, 9, 100, 100)
plt.imshow(bilateral_image, cmap='gray')
plt.show()
```

```python
dst_image = cv2.fastNlMeansDenoisingColored(image_RGB, None, 6, 6, 7, 21)
plt.imshow(dst_image, cmap='gray')
plt.show()
```

```python
thresh_image = cv2.adaptiveThreshold(input, 255, cv2.ADAPTIVE_THRESH_MEAN_C, 
                               cv2.THRESH_BINARY, 3, 5) 
plt.imshow(thresh_image, cmap='gray')
plt.show()
```

```python
_, th2_image = cv2.threshold(thresh_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(th2_image, cmap='gray')
plt.show()
```

```python
height, width = image_RGB.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 30, .75)
rotated_image = cv2.warpAffine(image_RGB, rotation_matrix, (width, height))
plt.imshow(rotated_image)
plt.show()
```

```python
rotated_image = cv2.transpose(image_RGB)
plt.imshow(rotated_image)
plt.show()
```

```python
flipped_image = cv2.flip(image_RGB, 0)
plt.imshow(flipped_image)
plt.show()
```

```python
image_scaled = cv2.resize(image_RGB, None, fx=0.75, fy=0.75)
plt.imshow(image_scaled)
plt.show()
```
