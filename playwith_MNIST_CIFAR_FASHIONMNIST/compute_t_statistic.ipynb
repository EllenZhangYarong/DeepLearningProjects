```python
def compute_mean_and_st(scores_of_sample_test):
    if(scores_of_sample_test.ndim == 2):
        mean_of_sample = np.mean(scores_of_sample_test[:,1])
        std_of_sample   = scores_of_sample_test[:,1].std()
    else:
        mean_of_sample = np.mean(scores_of_sample_test)
        std_of_sample   = scores_of_sample_test.std()
    nums_of_sample = len(scores_of_sample_test)  
    
    print("mean_of_sample = %.4f" % mean_of_sample)
    print("std_of_sample = %.4f" % std_of_sample)
    print("nums_of_sample = %d" % nums_of_sample)

    return mean_of_sample, std_of_sample, nums_of_sample
```

```python
def compute_t_statistic(array1,array2):
    mean_1,std_1,nums_1 = compute_mean_and_st(array1)
    mean_2,std_2,nums_2 = compute_mean_and_st(array2)
    SEM = np.sqrt(std_1*std_1/nums_1 + std_2*std_2/nums_2)
    t_statistic = (mean_2 - mean_1 )/SEM
    print("SEM = %.4f" % SEM)
    
    return t_statistic
```

```python
def look_up_t_table(alpha,df):
    
    if(30 <= df and df < 35 ):
        df = 30
        
    if(35 <= df and df < 45 ):
        df = 40
      
    if(45 <= df and df < 55 ):
        df = 50
        
    if(55 <= df and df < 80 ):
        df = 60
      
    if(80 <= df and df < 110 ):
        df = 100
        
    if(110 <= df and df < 500 ):
        df = 120
      
    if(500 < df ):
        df = 100000
        
    print(df)
      

    idx_alpha =  np.where(t_p[:] == alpha)
    idx_df    =  np.where(t_df[:] == df )
    t_critical = t_table[idx_df,idx_alpha]
    
    return t_critical[0,0]
```

```python
def compare_t_value(array1,array2,alpha,df):
    t_statistic = compute_t_statistic(array1,array2)
    t_critical  = look_up_t_table(alpha,df)
    
    print("t_statistic = %.4f" % t_statistic )
    print("t_critical = %f" % t_critical)
    print()
    if(t_statistic > t_critical):
#         Rsquare = t_statistic * t_statistic /(t_statistic * t_statistic + df)
#         print("Rsquare = %.2f%%" % Rsquare*100)
        print("Decision : Reject H0")
    else:
        print("Decision : Retain H0")
    
```
